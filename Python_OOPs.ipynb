23. How does polymorphism work in Python with inheritance?

Polymorphism with inheritance means that a child class can override methods of a parent class, and Python will decide which method to call at runtime based on the object type.

Example:
class Animal:
    def sound(self):
        return "Some sound"

class Dog(Animal):
    def sound(self):
        return "Bark"

class Cat(Animal):
    def sound(self):
        return "Meow"

def make_sound(animal):
    print(animal.sound())


Calling:

make_sound(Dog())   # Bark
make_sound(Cat())   # Meow


Polymorphism works because each subclass provides its own implementation of the parent method.

24. What is method chaining in Python OOP?

Method chaining means calling multiple methods on the same object in a single line, where each method returns the object (self).

Example:
class Number:
    def __init__(self, value):
        self.value = value

    def add(self, n):
        self.value += n
        return self    # enables chaining

    def multiply(self, n):
        self.value *= n
        return self

num = Number(5).add(3).multiply(2)
print(num.value)   # 16

Why it's useful:

Cleaner code

Fluent style

Useful in builders, data processing, ML pipelines

```python
#2.Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both
from abc import ABC, abstractmethod
import math

# Abstract Class
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

# Derived Class: Circle
class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius * self.radius


# Derived Class: Rectangle
class Rectangle(Shape):

    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def area(self):
        return self.length * self.breadth


# Testing the classes
circle = Circle(5)
rectangle = Rectangle(4, 6)

print("Area of Circle:", circle.area())
print("Area of Rectangle:", rectangle.area())
```

```python
#4 Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes Sparrow and Penguin that override the fly() method.
# Base Class
class Bird:
    def fly(self):
        return "Birds can usually fly."


# Derived Class 1
class Sparrow(Bird):
    def fly(self):
        return "Sparrow flies high in the sky."


# Derived Class 2
class Penguin(Bird):
    def fly(self):
        return "Penguins cannot fly."


# Function demonstrating polymorphism
def show_flying(bird):
    print(bird.fly())


# Creating objects
sparrow = Sparrow()
penguin = Penguin()

# Polymorphic behavior
show_flying(sparrow)   # Calls Sparrow's fly()
show_flying(penguin)   # Calls Penguin's fly()
```

```python
#5 Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.
class BankAccount:

    def __init__(self, initial_balance=0):
        self.__balance = initial_balance   # Private attribute

    # Method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: {amount}")
        else:
            print("Invalid deposit amount!")

    # Method to withdraw money
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew: {amount}")
        else:
            print("Insufficient balance or invalid amount!")
```

```python
#6 Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar and Piano that implement their own version of play().
# Base Class
class Instrument:
    def play(self):
        return "Instrument is playing sound."


# Derived Class 1
class Guitar(Instrument):
    def play(self):
        return "Guitar is playing melodious strings."


# Derived Class 2
class Piano(Instrument):
    def play(self):
        return "Piano is playing soft tunes."


# Function that demonstrates runtime polymorphism
def play_instrument(instrument):
    print(instrument.play())


# Creating objects of derived classes
g = Guitar()
p = Piano()

# Polymorphic behavior at runtime
play_instrument(g)   # Calls Guitar's play()
play_instrument(p)   # Calls Piano's play()
```

```python
#10 Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overloading the + operator
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"


# Testing the vector addition
v1 = Vector(2, 3)
v2 = Vector(4, 1)

v3 = v1 + v2   # Calls __add__()

print("Vector 1:", v1)
print("Vector 2:", v2)
print("Resultant Vector:", v3)
```

```python
#12 . Implement a class Student with attributes name and grades. Create a method average_grade() to compute the average of the grades
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades  # list of grades

    def average_grade(self):
        if len(self.grades) == 0:
            return 0
        return sum(self.grades) / len(self.grades)


# Testing the class
s1 = Student("Nimisha", [85, 90, 78, 92])

print("Student Name:", s1.name)
print("Average Grade:", s1.average_grade())
```

```python
#14 Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.
class Employee:
    def __init__(self, hours_worked, hourly_rate):
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate


class Manager(Employee):
    def __init__(self, hours_worked, hourly_rate, bonus):
        super().__init__(hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        base_salary = super().calculate_salary()
        return base_salary + self.bonus


# Testing the classes
emp = Employee(40, 200)
mgr = Manager(45, 300, 5000)

print("Employee Salary:", emp.calculate_salary())
print("Manager Salary:", mgr.calculate_salary())
```
